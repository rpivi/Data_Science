```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
delta = 0.00001
eps_vals=np.linspace(-delta, delta, 30)
alleigval=[]
allsingular=[]
for eps in eps_vals:
    A = np.array([[0, 1, 0, 0],[0, 0,2,0],[0,0,0,3],[eps,0,0,0]])
    U, S, Vh = np.linalg.svd(A, full_matrices=True)
    eigval,_=np.linalg.eig(A)
    alleigval.append(eigval)
    allsingular.append(S)
alleigval = np.array(alleigval)
allsingular=np.array(allsingular)

print(alleigval)
print(allsingular)

fig1, axs = plt.subplots(2, 2, figsize=(12, 10))
for i, ax in enumerate(axs.flat):
    ax.plot(eps_vals, alleigval[:, i].real, '-s', markersize=3)
    ax.plot(eps_vals, alleigval[:, i].imag, '-s', markersize=3)
    ax.set_title(f'Autovalore {i+1}')
    ax.set_xlabel("Perturbazione ")
    ax.set_ylabel("Valore")
    ax.grid(True)
fig1.legend(loc="upper right")
fig2, axs2 = plt.subplots(2, 2, figsize=(12, 10))
for i, ax in enumerate(axs2.flat):
    ax.plot(eps_vals, allsingular[:, i].real, '-s', markersize=3)
    ax.plot(eps_vals, allsingular[:, i].imag, '-s', markersize=3)
    ax.set_title(f'Valore Singolare {i+1}')
    ax.set_xlabel("Perturbazione ")
    ax.set_ylabel("Valore")
    ax.grid(True)
fig1.legend(["Re()", "Im()"], loc="upper right")
fig2.legend(["Re()", "Im()"], loc="upper right")
```
