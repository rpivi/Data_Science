```python
import numpy as np
import matplotlib.pyplot as plt
import cmath
```

```python
A = np.array([[-1, 1],[-1, -1]])

autovalori, autovettori= np.linalg.eig(A)

print(autovalori)
print(autovettori)
```

```python
D = np.array([[autovalori[0],0],[0, autovalori[1]]])

U = autovettori # è già la matrice con colonne gli autovettori


def eAt(t):
    Gt =  np.zeros((2,2),dtype=complex)
    for i in np.arange(2):
        Gt[i,i] = cmath.exp(autovalori[i]*t)
    return U @ Gt @ np.linalg.inv(U)
```

```python
t_vals = np.arange(0, 5.1, 0.1) 

eAt00r = []
eAt00i = []
eAt01r = []
eAt01i = []
eAt10r = []
eAt10i = []
eAt11r = []
eAt11i = []

for t in t_vals:
    eAt00r.append(eAt(t)[0,0].real)
    eAt00i.append(eAt(t)[0,0].imag)
    eAt01r.append(eAt(t)[0,1].real)
    eAt01i.append(eAt(t)[0,1].imag)
    eAt10r.append(eAt(t)[1,0].real)
    eAt10i.append(eAt(t)[1,0].imag)
    eAt11r.append(eAt(t)[1,1].real)
    eAt11i.append(eAt(t)[1,1].imag)

fig, axs = plt.subplots(2, 2, figsize=(12, 10))


axs[0,0].grid(True);
axs[1,0].grid(True);
axs[0,1].grid(True);
axs[1,1].grid(True);

axs[0, 0].plot(t_vals, eAt00r, '-o', markersize=1.5)
axs[0, 0].plot(t_vals, eAt00i,'-o', markersize=1.5)
axs[0, 0].set_title('Elemento [0,0]')
axs[0,0].set_xlabel("Parametro t")
axs[0,0].set_ylabel("Valore Elemento")

axs[0, 1].plot(t_vals, eAt01r, '-o', markersize=1.5)
axs[0, 1].plot(t_vals, eAt01i,'-o', markersize=1.5)
axs[0, 1].set_title('Elemento [0,1]')
axs[0,1].set_xlabel("Parametro t")
axs[0,1].set_ylabel("Valore Elemento")

axs[1, 0].plot(t_vals, eAt10r, '-o', markersize=1.5)
axs[1,0].plot(t_vals, eAt10i,'-o', markersize=1.5)
axs[1, 0].set_title('Elemento [1,0]')
axs[1,0].set_xlabel("Parametro t")
axs[1,0].set_ylabel("Valore Elemento")

axs[1, 1].plot(t_vals, eAt11r, '-o', markersize=1.5)
axs[1,1].plot(t_vals, eAt11i, '-o', markersize=1.5)
axs[1, 1].set_title('Elemento [1,1]')
axs[1,1].set_xlabel("Parametro t")
axs[1,1].set_ylabel("Valore Elemento")
fig.legend(["Re(λ)", "Im(λ)"], loc="upper right")
```
