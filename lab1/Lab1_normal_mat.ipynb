```python
import numpy as np
import matplotlib.pyplot as plt
import cmath
```

```python

Ns = np.array([2, 10, 50, 200])   # valori fissati di n
Nmat = 200              # numero di matrici generate per ciascun n

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

for ax, n in zip(axs.ravel(), Ns):
    alleig = []
    for i in range(Nmat):
        X = np.random.normal(loc=0.0, scale=1.0, size=(n, n))
        eigv = np.linalg.eigvals(X)
        scaled = eigv / np.sqrt(n)
        alleig.extend(scaled)

    alleig = np.array(alleig)

    ax.scatter(alleig.real, alleig.imag, s=0.2, alpha=0.8, color="purple")
    ax.set_title(f"Distribuzione autovalori - n = {n}")
    ax.set_xlabel("Parte Reale")
    ax.set_ylabel("Parte Immaginaria")
    ax.grid(True)

    #stesso range per tutti i plot
    ax.set_xlim([-2, 2])
    ax.set_ylim([-2, 2])

    circle = plt.Circle((0,0), 1, color="red", fill=False, lw=1)
    ax.add_artist(circle)
plt.show()
```
